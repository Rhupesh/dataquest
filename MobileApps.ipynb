```python
from csv import reader

## Apple Dataset ##
opened_data = open('AppleStore.csv')
read_data = reader(opened_data)
apple = list(read_data)
apple_header = apple[0]
apple = apple[1:]

## Google Dataset ##
opened_data = open('googleplaystore.csv')
read_data = reader(opened_data)
google = list(read_data)
google_header = google[0]
google = google[1:]
```

In below code, we explore the data by printing data

```python
def explore_data(dataset, start, end, rows_and_columns=False):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
```

```python
print(apple_header)
print('\n')
explore_data(apple, 0, 3, True)

print(google_header)
print('\n')
explore_data(google, 0, 3, True)
```

In below code, we are going to identify dulicate data and find a way to delete it. 

Instead of deleting it randomly, we will look at metrics to delete it

```python
unique_data = []
duplicate_data = []

for app in google:    
    name = app[0]
    if name in unique_data:
        duplicate_data.append(name)
    else:
        unique_data.append(name)

explore_data(duplicate_data, 0, 3, True)
```

In below code, we will remove duplicate entries based on number of reviews

```python
reviews_max = {}

for app in google:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    
    if name not in reviews_max:
        reviews_max[name] = n_reviews

print('Actual length:', len(reviews_max))
```

```python
android_clean = []
already_added = []

for app in google:
    name = app[0]
    n_reviews = float(app[3])
    
    if reviews_max[name] == n_reviews and name not in already_added:
        android_clean.append(app)
        already_added.append(name)

explore_data(android_clean, 0, 3, True)
```

In below codes, we will analyze apps with non english words and remove it

```python
def string_data(word):
    word_count = 0
    for character in word:
        if ord(character) > 127:
            word_count +=1
    
    if word_count > 3:
        return False
    else:
        return True

string_data('爱奇艺PPS -《欢乐颂2》电视剧热播')
```

```python
android_english = []
android_nonenglish = []

for app in android_clean:
    name = app[0]
    word_count = 0
    for character in name:
        if ord(character) > 127:
            word_count +=1  
    
    if word_count > 3:
        android_nonenglish.append(name)
    else:
        android_english.append(app)

explore_data(android_english, 0, 3, True)
print('Non English Count:',len(android_nonenglish))
```

```python
apple_english = []
apple_nonenglish = []

for app in apple:
    name = app[1]
    word_count = 0
    for character in name:
        if ord(character) > 127:
            word_count +=1  
    
    if word_count > 3:
        apple_nonenglish.append(name)
    else:
        apple_english.append(app)

print('Apple Apps Count:',len(apple))
explore_data(apple_english, 0, 3, True)
print('Non English Count:',len(apple_nonenglish))
print(apple_nonenglish)
```

Below code is the last step in cleaning process. We're going to remove apps which are not free.

```python
apple_free=[]
apple_notfree=[]

for app in apple_english:
    name = app[1]
    fee = app[4]         
    
    if fee == '0.0':
        apple_free.append(app)
    else:
        apple_notfree.append(name)
        
print('Apple Apps Count:',len(apple_english))
explore_data(apple_free, 0, 3, True)
print('Not Free Count:',len(apple_notfree))
```

```python
android_free=[]
android_notfree=[]

for app in android_english:
    name = app[0]
    fee = app[7]         
    
    if fee == '0':
        android_free.append(app)
    else:
        android_notfree.append(name)
        
print('Android Apps Count:',len(android_english))
explore_data(android_free, 0, 3, True)
print('Not Free Count:',len(android_notfree))
```

Now, the data cleaning is done. The company wants to find the genres where they can get more revenue from ads. Also, they are planning to implement this app in both android and apple marketplace.

In order to do this cost effective, a simple version of app will be built in android market and based on the success it will be implemented in apple market. So, in the data we're going to look for genres in both datasets.

```python
def freq_table(dataset,index):
    
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
    
    table_percentages = {}
    
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage 
    
    return table_percentages   

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
display_table(apple_free, 11)
```

In below code, we will find which app has most number of user ratings. From the results, I see higher number of user ratings for Navigation, Social Networking and Reference apps

```python
table = freq_table(apple_free, 11)

for key in table:
    genre = key
    total = 0
    len_genre = 0
        
    for app in apple_free:
        genre_app = app[11]
        if genre_app == genre:
            usr_rtng = float(app[5])
            total = total + usr_rtng
            len_genre += 1
            
    average = total/len_genre
    print(genre,':',average)
```

```python
table = freq_table(android_free, 1)

table_display = []

for key in table:
    category = key
    total = 0
    len_genre = 0
        
    for app in android_free:
        category_app = app[1]
        if category_app == category:
            n_installs = app[5]
            n_installs = n_installs.replace('+', '')
            n_installs = n_installs.replace(',', '')
            n_installs = float(n_installs)
            total = total + n_installs
            len_genre += 1
            
    average = total/len_genre
    print(category,':',average)
    key_val_as_tuple = (average, category)
    table_display.append(key_val_as_tuple) 
    
    
        
table_sorted = sorted(table_display, reverse = True)
for entry in table_sorted:
    print(entry[1], ':', entry[0])
```
