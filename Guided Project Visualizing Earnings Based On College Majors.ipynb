```python
raw_data_count = recent_grads.shape[0]
print("There are ", raw_data_count, "recent graduates in the data")

recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print("There are ", cleaned_data_count, "recent graduates in the cleaned data")
```
